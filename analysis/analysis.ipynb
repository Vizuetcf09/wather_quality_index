```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '../test/db/ind_ica_1.csv'
data = pd.read_csv(file_path)
data.info()
data.head()
data.describe()
```

```python
def plot_data(data, x_column, y_columns, number_columns=2, alpha=0.5):
    '''
    Grafica múltiples scatterplots en subplots.
    
    Parameters:
    - data: DataFrame
    - x_col: str, columna para eje X
    - y_cols: list, columnas para eje Y
    - ncols: int, número de columnas en la cuadrícula de subplots
    - alpha: float, transparencia de los puntos
    '''
    
    number_rows = -(-len(y_columns) // number_columns)  # redondear hacia arriba
    fig, axes = plt.subplots(number_rows, number_columns, figsize=(6 * number_columns, 4 * number_rows))
    axes = axes.flatten()  # Lista de ejes

    for i, y_column in enumerate(y_columns):
        ax = axes[i]
        ax.scatter(data[x_column], data[y_column], alpha=alpha)
        ax.set_xlabel(data[x_column])
        ax.set_ylabel(data[y_column])
        ax.set_title(f'{y_column} vs {x_column}')
        ax.set_xticklabels([])
        
    for j in range(len(y_columns), len(axes)):
        fig.delaxes(axes[j])
        
    plt.tight_layout()
    plt.show   
        
plot_data(data, x_column='NOMBRE_SITIO', y_columns=['DBO_TOT', 'DQO_TOT', 'COLI_FEC', 'E_COLI', 'OD_%', 'SST', 'group'], number_columns=2, alpha=0.5)
```
